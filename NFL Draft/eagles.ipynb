```python
import pandas as pd
```

```python
df = pd.read_csv("./data/eagles_roster.csv")
df.head()
```

```python
Drafted23 = df["Drafted (tm/rnd/yr)"].str.contains("2023", na=False)
Drafted22 = df["Drafted (tm/rnd/yr)"].str.contains("2022", na=False)

Drafted_22_or_23 = Drafted22 | Drafted23  # Correctly combine for "OR"

eagles_drafted = df["Drafted (tm/rnd/yr)"].str.contains(
    "Eagles", na=False
)  # Correctly apply str.contains to DataFrame column

eagles_drafted_combined_mask = (
    Drafted_22_or_23 & eagles_drafted
)  # Combine "Drafted in 22 or 23" AND "Eagles"

df_eagles_drafted = df[eagles_drafted_combined_mask]
df_eagles_drafted
```
