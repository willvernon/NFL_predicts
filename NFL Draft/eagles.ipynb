```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("./data/eagles_roster.csv")
df.head()
```

```python
Drafted23 = df["Drafted (tm/rnd/yr)"].str.contains("2023", na=False)
Drafted22 = df["Drafted (tm/rnd/yr)"].str.contains("2022", na=False)
Drafted24 = df[df["Yrs"] == 'Rook']

Drafted_22_or_23 = Drafted22 | Drafted23  # Correctly combine for "OR"

eagles_drafted = df["Drafted (tm/rnd/yr)"].str.contains(
    "Eagles", na=False
)  

eagles_drafted_combined_mask = (
    Drafted_22_or_23 & eagles_drafted
)  

df_eagles_drafted = df[eagles_drafted_combined_mask]
eagles_draft = pd.concat([df_eagles_drafted, Drafted24])
eagles_draft = eagles_draft.reset_index()
eagles_draft 
```

```python
def create_histogram(data, title="", x_label="", y_label="", color="skyblue", bins=2):
    plt.hist(data, bins=bins, color=color, edgecolor='black')
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.grid(axis='x', alpha=0.5)
    plt.show()
```

```python
eagles_draft.loc[eagles_draft["Yrs"] == 'Rook', 'Yrs'] = '0'
```

```python
eagles_draft_av = eagles_draft.groupby(by='Yrs', sort=True, dropna=True)['AV'].sum().reset_index()
eagles_draft_av
```

# Draftee AV 22',23',24'

The sum of AV for the Draft Class still on the roster:

```python
plt.bar(eagles_draft_av['Yrs'], eagles_draft_av['AV'] )
plt.xlabel('Yrs')
plt.ylabel('AV')
plt.title('Draftees 22-24')
plt.show()
```
