```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
np.random.seed(0)
x = np.random.normal(loc=0.0, scale=1.0, size=1000)  # Normally distributed data
y = np.random.exponential(
    scale=1.0, size=1000
)  # Exponentially distributed data for comparison

# Compute the quantiles
quantiles_x = np.sort(x)
quantiles_y = np.sort(y)
probs = np.linspace(0, 1, len(quantiles_x))

# Compute theoretical quantiles from a normal distribution
# for comparison with both x and y
normal_quantiles = stats.norm.ppf(probs, loc=0, scale=1)

# Plot
plt.figure(figsize=(10, 6))

# Plot for x (which should align with normal quantiles if x is normally distributed)
plt.plot(normal_quantiles, quantiles_x, "b.", alpha=0.5, label="Sample vs Normal (x)")

# Plot for y (to show deviation from normality)
plt.plot(normal_quantiles, quantiles_y, "r.", alpha=0.5, label="Sample vs Normal (y)")

# Add 45-degree line for reference
plt.plot(normal_quantiles, normal_quantiles, "k-", lw=2, label="Normal Line")

plt.title("Q-Q Plot")
plt.xlabel("Theoretical Quantiles")
plt.ylabel("Sample Quantiles")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
